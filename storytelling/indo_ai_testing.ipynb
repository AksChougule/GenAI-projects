```python
import requests
```

```python
headers = {"Content-Type": "application/json",
           'API-Subscription-Key': 'enter your API key'}
```

```python
text_to_translate = """1. A language is something infinitely greater than grammar and philology. It is the poetic testament of the genius of a race and a culture, and the living embodiment of the thoughts and fancies that have moulded them.
2. India has known the innocence and insouciance of childhood, the  passion and abandon of youth, and the ripe wisdom of maturity that comes from long experience of pain and pleasure; and over and over a gain she has renewed her childhood and youth and age”
  ― Jawaharlal Nehru, The Discovery of India"""
```

```python
url = "https://api.sarvam.ai/translate"

payload = {
    "inputs": [text_to_translate],
    "source_language_code": "en-IN",
    "target_language_code": "mr-IN",
    "mode": "formal",
    "model": "mayura:v1",
    "enable_preprocessing": True
}
```

```python
response = requests.request("POST", url, json=payload, headers=headers)
```

```python
print(response.text)
```

```python
translated_text = response.text.replace('{"translations":["','').replace('"]}','').replace('\\n','')
translated_text 
```

## Text to Audio, approach 2

```python
import requests

url = "https://api.sarvam.ai/text-to-speech"

payload = {
    "inputs": [translated_text],
    "target_language_code": "mr-IN",
    "speaker": "meera",
    "pitch": 0,
    "pace": 1.65,
    "loudness": 2,
    "speech_sample_rate": 8000,
    "enable_preprocessing": True,
    "model": "bulbul:v1"
}
```

```python
response = requests.request("POST", url, json=payload, headers=headers)

print(response.text)
```

```python
type(response)
```

```python
import base64
from IPython.display import Audio
```

```python
audio_string = response.text[12:-3]
audio_data = base64.b64decode(audio_string)

with open("t2s.wav","wb") as audio_file:
    audio_file.write(audio_data)
```

```python
audio_file = "t2s.wav"
Audio(audio_file)
```

```python
# the clip is too fast, let try to change the speaker and slow is down

payload = {
    "inputs": [translated_text],
    "target_language_code": "mr-IN",
    "speaker": "arvind",
    "pitch": 0,
    "pace": 1.15,
    "loudness": 2,
    "speech_sample_rate": 8000,
    "enable_preprocessing": True,
    "model": "bulbul:v1"
}

response = requests.request("POST", url, json=payload, headers=headers)

audio_string = response.text[12:-3]
audio_data = base64.b64decode(audio_string)

with open("t2s_v2.wav","wb") as audio_file:
    audio_file.write(audio_data)

audio_file = "t2s_v2.wav"
Audio(audio_file)
```
